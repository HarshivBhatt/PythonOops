### Small Banking Program use case designed with the help of Oops concept
#### Here i have created one solution in which i tried to cover the daily banking operation like Opening account, add amount in the account, withdraw money from the account, know the balance of account etc.
#### Technical concept i have covered: Oops mechanisam, Abstract class, Inheritance, Use of super keyword


```python
#importing all the statement which is further used in program
from abc import ABC, abstractmethod
from random import randint
```

```python
#Here i m creating one Abstract class 'Account' in which i defined some methods.
class Account:
    #This class contains several abstract method. All the abstract methods is defined with the decorator of abstractmethod. 
    @abstractmethod
    def createAccount():
        return 0
    @abstractmethod
    def authenticate():
        return 0
    @abstractmethod
    def withdraw():
        return 0
    @abstractmethod          
    def deposite():
        return 0
    @abstractmethod
    def displaybalance():
        return 0

# Now i create second class 'SavingsAccount', which inherit above abstract class.
class SavingAccount(Account):
    #define constructor with the class variable savingAccounts. 
    def __init__(self):
        #Here we defined savingAccounts as a dictionary in which we store the name of account holder,
#       #account number and intial deposite
        self.savingAccounts = {"11111" : ["hr",1]}
    
    def createAccount(self, name, initialDeposite):
        '''This function takes name & intialdeposite as an input and store it to the class variable created above'''
        self.accountNumber = randint(10000,99999)
        self.savingAccounts[self.accountNumber] = [name, initialDeposite]
        print("Your account is successfully created, Your account number is {}".format(self.accountNumber))
        
    def authenticate(self, name, accountNumber):
        '''This function verify the identity of user so for that it takes name and account number as an input and returns
        boolean value True if it authenticate successfully, otherwise it will return false'''
        print(accountNumber)
        print(type(accountNumber))
        print(self.savingAccounts.keys())
        if(accountNumber in self.savingAccounts.keys()):
            if(self.savingAccounts[accountNumber][0] == name):
                print("Authentication Successful!")
                self.accountNumber = accountNumber
            return True
        else:
            print("all keys >>>>>>>>>>>>>" + str(self.savingAccounts.keys()))
            print("name you enter>>>>>>>>>" + name) 
#             print("name in database>>>>>>>" + self.savingAccounts[str(accountNumber)][0]) 
            print("Authentication failed") 
            return False

    def withdraw(self, withdrawAmount):
        '''This function takes the withdraw amount as an input and perform the calculation on exsiting account balance'''
        if(withdrawAmount > self.savingAccounts[self.accountNumber][1]):
            print("Insufficient Balance") 
        else:
            self.savingAccounts[self.accountNumber][1] -= withdrawAmount
            print("Withdraw Successful") 
            self.displaybalance(self.accountNumber)

    def deposite(self, depositeAmount):
        '''This function takes deposite amount as an input and perform addition on the exsiting account balance'''
        self.savingAccounts[self.accountNumber][1] += depositeAmount
        print("deposite Successful") 
        self.displaybalance(self.accountNumber)

    def displaybalance(self, accountNumber):
        '''This function display the account balance'''
        print("Available balance: {}".format(self.savingAccounts[accountNumber][1])) 

###inherited SavingAccount class into this class. 
class Add_acc_holder(SavingAccount):
    '''This class works when first account holder wants to add second name into the account'''
    def __init__(self,first_name,second_name,verify_acc):
        self.first_name = first_name
        self.second_name = second_name
        self.verify_acc = verify_acc
        #Here super() method invoke all the properties of Parent class:SavingAccount into this class
        super().__init__()
#         super().__init__(name,accountNumber)

    def add_second_name(self,first_name,verify_acc):
        '''This function add the second name of account holder after it verify all the details'''
        auth_result = super().authenticate(self.first_name,self.verify_acc)
        if auth_result == 'True':
            print('New name added Successfully')
            print(self.name,self.second_name)
        else:
            print('Sorry, please verify your account number!')

#create one object of SavingAccount class.
savingAccount = SavingAccount()
while True:
    print("Enter 1 to open an acc")
    print("Enter 2 to acess existing account") 
    print("Enter 3 to Add second account holder name")
    print("Enter 4 to exit")
    #Here we take the input from user.like which operation you want to do
    userChoice = int(input(""))
    if(userChoice == 1):
        name = input("Enter your name: ")
        initialDeposite = int(input("Enter initialDeposite"))
        #calling createAccount mehtod and it stored the value that has been entered 
        savingAccount.createAccount(name,initialDeposite)

    elif(userChoice == 2):
        name = input("Enter name: ")
        accountNumber = int(input("Enter accountNumber: "))
        #calling authenticate mehtod and it verify the identity of user
        authenticateStatus = savingAccount.authenticate(name, accountNumber)
        if(authenticateStatus is True):
            while True:
                #If the status is True means successfully verified the user details, then user can do further operation
                print ("enter 1 to withdraw")
                print("Enter 2 to deposite") 
                print("Enter 3 to displaybalance") 
                print("Enter 4 to exit")
                userChoice = int(input(""))
                if(userChoice == 1):
                    withdrawAmount = int(input("Enter withdrawAmount : "))
                    savingAccount.withdraw(withdrawAmount)
                elif(userChoice == 2):
                    depositeAmount = int(input("Enter depositeAmount : "))
                    savingAccount.deposite(depositeAmount)
                elif(userChoice == 3):
                    savingAccount.displaybalance(accountNumber)
                elif(userChoice == 4):
                    break
    elif(userChoice == 3):
        first_name = input('Enter first name: ')
        acc_number = int(input('Enter account number: '))
        second_name = input("Enter second name: ")
        #Here we pass the value for adding the second name into the account
        add_acc_name = Add_acc_holder(first_name,second_name,acc_number)
        #Calling add_second_name method for adding name. 
        add_acc_name.add_second_name(first_name,acc_number)
    elif (userChoice == 4):
        print('Thank you!')
        break           
```
